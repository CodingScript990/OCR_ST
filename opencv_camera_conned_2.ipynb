```python
# 카메라 연결 코드

# 개발환경

# 1. Python(프로그램 언어)
# 2. vscode(개발도구)
# 3. LENOVO 300 FHD(카메라)

import cv2 # openCV 2 라이브러리(영상 처리를 위한 것)
import datetime # datetime 시간에 관련된 모듈이다
import numpy as np # numpy 라이브러리(계산을 위한 것)
from PIL import ImageFont, ImageDraw, Image # PIL 이미지 분석 및 처리를 쉽게 할 수 있는 라이브러리

# Pillow - 영상에 글자를 입히는 모듈이다
# PIL은 opencv로 image를 read할 경우 이미지를 PIL로 변환해주는 라이브러리이다
```

```python
# openCV Python
# 카메라 영상을 받아올 개게 선언 및 설정(영상 소스, 해상도 설정)

cap = cv2.VideoCapture(0) # cap이라는 객체를 선언 후, 0번째 웹캠으로부터 영상의 스냅샷을 가져온다는 뜻이다(0은 default)
# VideoCapture Class 인스턴스를 생성 후

# 해상도 설정

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640) # 너비

cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480) # 높이

# 영상을 기록할 코덱 설정(어떻게 인코딩을 할 것인가?)
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# 글꼴 설정

font_size = 20 # 글꼴 사이즈

font = ImageFont.truetype('arial.ttf', font_size) # arial.ttf 글씨체

# 녹화상태는 처음에는 거짓으로 설정한다
is_record = False # 녹화상태 구현하기 위함이다
```

```python
# 무한루프

while True:
    
    now_DT = datetime.datetime.now() # now 현재 시간을 보여주는 함수
    
    now_Datetime = now_DT.strftime('%Y-%m-%d %H:%M:%S') # 현재시각을 불러와 문자열로 저장한다
    
    now_Datetime_path = now_DT.strftime('%Y-%m-%d %H_%M_%S') # 파일이름으로는 :을 사용할 수 없기에 따로 만들어준다
    
    ret, frame = cap.read() # 그리고 cap객체에서 0번째 영상의 스냅샷을 읽는다
    
    # 글자가 잘보이도록 배경색을 넣어준다
    cv2.rectangle(img=frame, pt1=(10, 15), pt2=(245, 35), color=(0,0,0), thickness=-1)
    # img는 사각형을 넣을 이미지, pt1, pt2는 사각형의 시작점과 끝점이고, color는 색상(파랑, 초록, 빨강), tickness는 선굵기(-1:내부를 채움)
    
    # 아래의 4줄은 글자를 영상에 더해주는 역할을 한다
    
    frame = Image.fromarray(frame) # 이미지를 numpy배열을 PIL Image로 변환한다
    
    draw = ImageDraw.Draw(frame) # 이미지를 드로우한다
    
    # xy는 텍스트 시작위치이고, text는 출력할 문자열을 말하고, fill은 글자색(파랑, 초록, 빨강)을 말한다
    draw.text(xy=(10,15), text='Test ' + now_Datetime, font=font, fill=(255, 255, 255))
    
    frame = np.array(frame) # 현재시간을 표시하는 글자를 써준 영상을 출력한다
    
    #cv2.imshow('iVCam', frame) # frame(카메라 영상)을 iVCam이라는 창에 띄워준다(이미지 화면에 표시)
    
    # 30ms 동안 키입력 대기
    key = cv2.waitKey(3)
    
    # if(만약) 현재상태가 녹화중이 아니라면? r키를 누르면?
    if key == ord('r') and is_record == False:
        
        is_record = True # 녹화상태로 만들어준다
        
        # 비디오 객체에 파일이름(한글가능)하고, 인코더이름, 초당 프레임률(정확하지 않다), 영상크기로 영상을 쓸 준비한다
        video = cv2.VideoWriter('Test ' + now_Datetime_path + '.avi', fourcc, 15, (frame.shape[1], frame.shape[0]))
    
    # 녹화중인 상태에서 다시 r키를 누르면?
    elif key == ord('r') and is_record == True:
        
        is_record = False # 녹화상태를 꺼준다
        
        video.release() # 녹화를 종료한다
    
    # 녹화중인 상태에서 다시 c키를 누르면?
    elif key == ord('c'):
        # 파일이름(한글불가, 영어만 가능), 이미지로 영상을 캡처하여 그림파일로 저장한다
        cv2.imwrite('cap' + now_Datetime_path + '.bmp', frame) # 파일이름(한글X), 이미지
        
    # 녹화중인 상태에서 다시 q를 누르면 무한루프가 멈춘다
    elif key == ord('q'):
        
        break # 멈춘다
    
    # 현재상태가 녹화상태이면?
    if is_record == True:
        
        # 비디오 객체에 현재 프레임을 저장한다
        video.write(frame)
        
        # 녹화중이라는 것을 보여주기 위해 화면에 빨간색 점으로 표시해준다
        cv2.circle(img=frame, center=(620, 15), radius=5, color=(0,0,255), thickness=-1)
        
    # 현재 시간을 표시하는 글자를 써준 영상을 출력한다
    cv2.imshow('output', frame)
    
    # if(만약) cv2라이브러리.waitKey(1초)후, ord(유니코드 값을 돌려주는 함수)
    if cv2.waitKey(1) == ord('q'): # 키보드의 q를 누르면 무한루프가 멈춘다
        break # stop
    
cap.release() # 스냅샷 객체를 없애준다(release() => 메서드)
cv2.destroyAllWindows() # 모든 영상 창을 닫아준다(이미지 윈도우 삭제)
```
